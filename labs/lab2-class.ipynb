```python
import pandas as pd
import numpy as np

#data from wc_day6_1_sample.csv
df = pd.read_csv("wc_day6_1_sample.csv", encoding="ISO-8859-1", names=["ClientID", "Date","Time","URL", "ResponseCode","Size"])
                 #header=-1) 
                 #names=['index', 'date', 'time', 'file', 'status', 'duration'])
df['Size'] = df['Size'].replace('-', np.nan)
df['Size'] = df['Size'].astype(float)
```

```python
url = df[df['ResponseCode'] != "200"]
url
```

```python
group = df.groupby('ResponseCode')
group.groups.keys()
```

```python
group.get_group('200')
group.size()
```

```python
import matplotlib as plt
% matplotlib inline
#group.size().plot(kind='pie', legend=True)
ax = group.size().plot(kind='bar')
ax.set_yscale('log', nonposy='clip')
```

```python
may1_df = df[df['Date'] == '01/May/1998']
x = may1_df.apply(lambda x: ' '.join([x['Date'], x['Time']]), axis=1)
may1_df.loc[:,('Datetime')] = pd.to_datetime(x)
may1_df
```

```python
hg = may1_df.groupby(lambda x: may1_df['Datetime'][x].hour)
```

```python
hg.groups.keys()
```

```python
hg['Size'].sum()
```

```python
ax = hg.size().plot(kind="bar")
ax.set_ylabel("# Requests")
ax.set_xlabel("Hour of the day")
ax.set_title("# Request changes in a day")
```

```python
fig, ax1 = plt.pyplot.subplots()
ax2 = ax1.twinx()
x = hg.size().index

ax1.plot(x, hg.size(), 'b-')
ax2.plot(x, hg['Size'].sum(), 'r-')

ax1.set_xlabel('Hour of the day')
ax1.set_ylabel('# Requests', color='b')
ax2.set_ylabel('Size of traffic handled', color='r')
```
